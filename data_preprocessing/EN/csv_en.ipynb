```python
import pandas as pd
import shutil
#Читаю обучающий файл  
train=pd.read_csv(r'D:\data_en\en\en\en\train.tsv',  sep="\t", encoding="utf-8")
```

```python
train
```

```python
#удаляю ненужные столбцы
del train['sentence']
del train['up_votes']
del train['down_votes']
del train['locale']
del train['segment']
del train['accent']
del train['client_id']
```

```python
train
```

```python
#теперь необходимо избавиться от строк с gender == Nan и age == NaN
train = train.dropna(axis = 0, how = 'any')
train = train[train["gender"] != 'other']

train = train.dropna(axis = 0, how = 'any')
train = train[train["age"] != 'other']
```

```python
train
```

```python
#сохраняем csv - файл для нашего нового обучающего датасета 
train.to_csv('D:/data_en/train_en.csv', index=False, sep="\t")
```
