```python
import os
import pandas as pd
import shutil
#Читаю обучающий файл 
train=pd.read_csv(r'D:\dataset_mp3\a_cvs_info\train.csv',  sep="\t", encoding="utf-8")
#Вывожу таблицу
train_labels=train[["path", "age", "gender"]]
train_labels
```

```python
male_train = train_labels[train_labels["gender"]=="male"] #выбираю файлы в train.csv, которые имеют gender == male
female_train = train_labels[train_labels["gender"]=="female"]
fifties_train = train_labels[train_labels["age"]=="fifties"]
sixties_train = train_labels[train_labels["age"]=="sixties"]
thirties_train = train_labels[train_labels["age"]=="thirties"]
twenties_train = train_labels[train_labels["age"]=="twenties"]
teens_train = train_labels[train_labels["age"]=="teens"]
fourties_train = train_labels[train_labels["age"]=="fourties"]
```

```python
paths = os.listdir('D:/image')
#перекидываю из общей папки изображения в директорию train/male 
for f in male_train["path"].tolist():
  f = f.replace('.mp3', '.wav.png')
  if f in paths:
    shutil.copy('D:/image/' + f, "D:/train/male/"+ f)
paths = os.listdir('D:/image') 


#теперь train >> train/female
for f in female_train["path"].tolist():
  f = f.replace('.mp3', '.wav.png')
  if f in paths:
    shutil.copy('D:/image/' + f, "D:/train/female/"+ f)
    
    
# train >> fifties
for f in fifties_train["path"].tolist():
  f = f.replace('.mp3', '.wav.png')
  if f in paths:
    shutil.copy('D:/image/' + f, "D:/train/fifties/"+ f)
    

# train >> sixties
for f in sixties_train["path"].tolist():
  f = f.replace('.mp3', '.wav.png')
  if f in paths:
    shutil.copy('D:/image/' + f, "D:/train/sixties/"+ f)
    

# train >> thirties
for f in thirties_train["path"].tolist():
  f = f.replace('.mp3', '.wav.png')
  if f in paths:
    shutil.copy('D:/image/' + f, "D:/train/thirties/"+ f)
    

# train >> twenties
for f in twenties_train["path"].tolist():
  f = f.replace('.mp3', '.wav.png')
  if f in paths:
    shutil.copy('D:/image/' + f, "D:/train/twenties/"+ f)
    

# train >> teens
for f in teens_train["path"].tolist():
  f = f.replace('.mp3', '.wav.png')
  if f in paths:
    shutil.copy('D:/image/' + f, "D:/train/teens/"+ f)
    

# train >> fourties
for f in fourties_train["path"].tolist():
  f = f.replace('.mp3', '.wav.png')
  if f in paths:
    shutil.copy('D:/image/' + f, "D:/train/fourties/"+ f)
```
